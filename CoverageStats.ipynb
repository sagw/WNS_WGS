```r
none=coverage[coverage$depth==0 & coverage$percentage==100,]
```

```r
nrow(none)
```

```r
sort(table(droplevels(none$Chromosome)))
```

```r
noneAll=coverage[which(coverage$Chromosome %in% none$Chromosome),]
```

```r
nrow(noneAll) 
```

```r
noneAllZero=noneAll[noneAll$depth==0,]
```

```r
nrow(noneAllZero)
```

```r
head(noneAllZero)
```

```r
noneAllZero$percentageCov=(100-(noneAllZero$percentage))
```

```r
head(noneAllZero)
```

```r
require(plyr)
```

```r
require(ggplot2)
```

```r
noneAllZero=noneAllZero[order(noneAllZero$Chromosome),] 
```

```r
noneAllzeroMean=ddply(noneAllZero, c("Chromosome"), summarise,mean = mean(percentageCov))
```

```r
head(noneAllzeroMean)
```

```r
noneAllzeroMean=noneAllzeroMean[order(noneAllzeroMean$mean),] 
```

```r
head(noneAllzeroMean)
```

```r
noneAllZero=noneAllZero[order(match(noneAllZero$Chromosome, noneAllzeroMean$Chromosome)),] 
```

```r
head(noneAllZero)
```

```r
noneAllZero$Chromosome=factor(noneAllZero$Chromosome, levels=unique(as.character(noneAllZero$Chromosome)) )
```

```r
write.csv(noneAllZero, file="Scaffold1xCov.csv")
```

```r
plot=ggplot(data=noneAllZero, aes(x=Chromosome, y=percentageCov))+
  geom_point(data=noneAllZero, stat="identity", aes(color=sample))+
  theme_bw()+ylab("Percent scaffold >= 1x coverage")+
xlab("Scaffold")+
  theme(axis.text.x = NULL)
```

```r
plot
```

```r
zero=coverage[coverage$depth==0,]
```

```r
head(zero)
```

```r
nrow(zero)
```

```r
zero$percentage=NULL
```

```r
zero$percent=NULL
```

```r
zero$depth=NULL
```

```r
head(zero)
```

```r
require(plyr)
```

```r
zero$bpCovered=as.numeric(as.character(zero$bpCovered))
```

```r
zero$bpTotal=as.numeric(as.character(zero$bpTotal))
```

```r
zerosum=ddply(zero, "sample", numcolwise(sum))
```

```r
head(zerosum)
```

```r
zerosum$percentage=(1-(zerosum$bpCovered/zerosum$bpTotal))*100
```

```r
zerosum$EU=c("E", "E", "E","E","E","E","E","E","U","U","U","U","U","U","U","U","U","U","U","U","U","U")
```

```r
zerosum$diluted=c("YES", "YES", "YES","YES","YES","YES","YES","YES","YES","NO","NO","NO","NO","YES", "NO","YES", "NO","NO","YES","NO","NO","YES")
```

```r
zerosum
```

```r
require(ggplot2)
```

```r
plot=ggplot(data=zerosum, aes(x=sample, y=percentage))+
  geom_bar(data=zerosum, stat="identity", aes(color=diluted, fill=EU))+
  theme_bw()+ylab("Percent >= 1x coverage")+scale_color_manual(values=c("darkblue", "red"))+
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```
