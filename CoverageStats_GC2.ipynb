```r
Cov=read.csv("Scaffold1xCov.csv", header=TRUE)
```

```r
Cov=apply(Cov, 2, function(y) gsub('_','',y))
```

```r
Cov2=subset(Cov, select=c("Chromosome", "percentageCov", "sample"))
```

```r
colnames(Cov2)=c("Scaffold", "PercentCov", "Sample")
```

```r
colnames(Cov2)
```

```r
colnames(Samples)
```

```r
CovSamp=merge(Cov2, Samples, by=c( "Scaffold","Sample"))
```

```r
CovSamp
```

```r
CovSamp$PercentCov=as.numeric(as.character(CovSamp$PercentCov))
```

```r
CovSamp
```
